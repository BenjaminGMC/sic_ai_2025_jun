```python
features = [
    "study_hours_per_week",
    "sleep_hours_per_day",
    "attendance_percentage",
    "assignments_completed",
    "participation_level_Low",
    "participation_level_Medium",
    "internet_access_Yes",
    "parental_education_High School",
    "parental_education_Master's",
    "parental_education_PhD",
    "extracurricular_Yes",
    "part_time_job_Yes"
]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

# Determinar n√∫mero √≥ptimo de clusters con el m√©todo del codo
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.title("M√©todo del Codo")
plt.xlabel("N√∫mero de Clusters")
plt.ylabel("Inercia")
plt.grid(True)
plt.show()
```

```python
# Metodo Silhouette
from sklearn.metrics import silhouette_score

silhouette_scores = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

# 4. Graficar
plt.plot(K_range, silhouette_scores, marker='o')
plt.xlabel("N√∫mero de clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("M√©todo del Silhouette para determinar k")
plt.grid(True)
plt.show()
```

K-Means con 3 clusters

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # cambia el n√∫mero seg√∫n el codo
df["cluster"] = kmeans.fit_predict(X_scaled)
```

```python
cluster_summary = df.groupby("cluster")[features + ["final_grade"]].mean()
display(cluster_summary)

sns.boxplot(data=df, x="cluster", y="final_grade")
plt.title("Distribuci√≥n de notas por cluster")
plt.show()
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Reducci√≥n de dimensionalidad para visualizaci√≥n
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# A√±adir a df
df["pca1"] = X_pca[:, 0]
df["pca2"] = X_pca[:, 1]

# Transformar los centroides al mismo espacio PCA
centroids_pca = pca.transform(kmeans.cluster_centers_)

# Graficar
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="pca1", y="pca2", hue="cluster", palette="Set2", alpha=0.7, s=50)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c='black', marker='X', s=200, label='Centroides')
plt.title("Visualizaci√≥n de clusters con centroides (PCA)")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear df_prototypes excluyendo las columnas de educaci√≥n parental
df_prototypes = df.drop(columns=[
    "parental_education_High School",
    "parental_education_Master's",
    "parental_education_PhD",
    "part_time_job_Yes"
])
```

```python
# Invertir los one-hot encoding para columnas categ√≥ricas
def revert_dummies(df_prototypes, col_prefixes):
    for prefix in col_prefixes:
        cols = [col for col in df_prototypes.columns if col.startswith(prefix)]
        df_prototypes[prefix] = df_prototypes[cols].idxmax(axis=1).str.replace(prefix + "_", "")
        df_prototypes.drop(columns=cols, inplace=True)
    return df_prototypes
```

```python
# Revertir dummies en las columnas categ√≥ricas
cat_prefixes = ["participation_level", "internet_access", "extracurricular"]
df_clean = revert_dummies(df_prototypes.copy(), cat_prefixes)
```

```python
# Convertir todas las columnas categ√≥ricas a string
for col in cat_prefixes:
    df_clean[col] = df_clean[col].astype(str)

# Convertir a numpy array
X = df_clean.to_numpy()

# Obtener √≠ndices de columnas categ√≥ricas
cat_cols_idx = [df_clean.columns.get_loc(col) for col in cat_prefixes]
```

```python
df_clean.head()
```

```python
display(X)
```

```python
# Aplicar K-Prototypes
kproto = KPrototypes(n_clusters=3, init='Huang', verbose=1, random_state=42)
clusters = kproto.fit_predict(X, categorical=cat_cols_idx)
```

```python
# Agregar clusters al DataFrame
df_clean["Cluster"] = clusters

# Ver resultado
display(df_clean.groupby("Cluster").mean(numeric_only=True))
display(df_clean["Cluster"].value_counts())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

# Escalar solo las columnas num√©ricas
num_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df_clean[num_cols])

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_clean["pca1"] = X_pca[:, 0]
df_clean["pca2"] = X_pca[:, 1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aseg√∫rate de que 'cluster' sea tipo entero para que los colores se asignen correctamente
df_clean['cluster'] = df_clean['Cluster'].astype(int)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='pca1', y='pca2', hue='cluster', palette='Set1', s=80, alpha=0.7)

plt.title("Visualizaci√≥n de Clusters con K-Prototypes (PCA)", fontsize=14)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import make_blobs, make_moons
import matplotlib.pyplot as plt
```

```python
x1, label1 = make_blobs(n_samples=200, n_features=2, centers=2, cluster_std=5, random_state=123)
plt.scatter(x1[:, 0], x1[:, 1], c=label1, alpha = 0.7)
plt.title("dataset : ej 01")
plt.show()
```

```python
x2, label2 = make_moons(n_samples=200, noise=0.08, random_state=123)
plt.scatter(x2[:, 0], x2[:, 1], c=label2, alpha = 0.7)
plt.title("dataset : ej 01")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
```

```python
x1df = pd.DataFrame(x1)
x1df.head()
```

```python
x2df = pd.DataFrame(x2)
x2df.head()
```

```python
aglo = AgglomerativeClustering(n_clusters=2)
aglo.fit(x1)
```

```python
myColors = {0: 'red', 1: 'blue'}
plt.scatter(x1[:,0], x1[:,1], c = pd.Series(aglo.labels_).apply(lambda x: myColors[x]), alpha = 0.7)
plt.title("dataset : agglomerative")
plt.show()
```

```python
aglo = AgglomerativeClustering(n_clusters=4) #revisar otros tipos de aglomeracion
aglo.fit(x2)

myColors = {0: 'red', 1: 'blue', 2: 'green', 3: 'yellow'}
plt.scatter(x2[:,0], x2[:,1], c = pd.Series(aglo.labels_).apply(lambda x: myColors[x]), alpha = 0.7)
plt.title("dataset : agglomerative")
plt.show()
```

```python
from sklearn.metrics import pairwise_distances
def plot_clusters(X, labels, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.title(title)
    plt.show()
D = pairwise_distances(x2, metric='cosine')

model = AgglomerativeClustering(n_clusters=3, linkage='average', metric='precomputed')
labels = model.fit_predict(D)
plot_clusters(x2, labels, 'linkage=average, metric=precomputed (cosine)')
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
myLinkage = linkage(x1, method="single")
plt.figure(figsize=(20, 5))
dendrogram(myLinkage)
plt.show()
```

```python
labels = fcluster(myLinkage, 5, criterion='distance')
pd.Series(labels).value_counts()
```

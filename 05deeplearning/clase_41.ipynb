# Resumen Clase 41 – Series de Tiempo

## 1. Introducción
- Las **series de tiempo** son datos organizados cronológicamente.
- Se usan para analizar el pasado y **predecir valores futuros**.
- Aplicaciones: economía, finanzas, clima, producción industrial, salud, entre otras.

---

## 2. Componentes de una Serie de Tiempo
1. **Tendencia (T):** evolución a largo plazo.
2. **Estacionalidad (E):** patrones que se repiten en intervalos regulares (ej: meses, estaciones).
3. **Ciclo (C):** fluctuaciones más largas, relacionadas con fenómenos económicos o sociales.
4. **Ruido (R):** variaciones aleatorias no explicadas.

---

## 3. Métodos de Análisis
- **Descomposición aditiva y multiplicativa**:
  - Aditiva: Serie = Tendencia + Estacionalidad + Ruido.
  - Multiplicativa: Serie = Tendencia × Estacionalidad × Ruido.
- **Suavizamiento**:
  - Medias móviles.
  - Suavizamiento exponencial simple y Holt-Winters.
- **Modelos clásicos**:
  - AR (Autoregresivo).
  - MA (Media móvil).
  - ARMA y ARIMA (integrado, útil para series no estacionarias).

---

## 4. Estacionariedad
- Una serie es **estacionaria** si sus propiedades estadísticas no cambian con el tiempo.
- Es requisito para la mayoría de los modelos ARIMA.
- Se detecta con:
  - Visualización de la serie.
  - Tests estadísticos (ej: Dickey-Fuller).
- Se logra con **diferenciación** o transformaciones.

---

## 5. Ejemplo en R
- Uso del dataset `AirPassengers`.
- Pasos trabajados:
  - Cargar la serie.
  - Visualizar para identificar tendencia y estacionalidad.
  - Descomposición de la serie.
  - Aplicación de suavizamiento exponencial.

---

## 6. Conclusiones
- El análisis de series de tiempo permite **capturar patrones temporales**.
- Es clave diferenciar entre tendencia, estacionalidad y ruido.
- Preparación para modelos más avanzados:
  - ARIMA.
  - Modelos con Machine Learning.
  - Aplicaciones con datos reales.


## Ejemplo en r

```r
install.packages("forecast")
```

```r
# Cargar librería
library(forecast)

# Dataset de ejemplo
data("AirPassengers")
ts_data <- AirPassengers

# Visualización
plot(ts_data, main="Pasajeros de Aerolíneas (1949-1960)",
     xlab="Año", ylab="Número de pasajeros")

# Descomposición
decomp <- decompose(ts_data)
plot(decomp)

# Suavizamiento exponencial Holt-Winters
model <- HoltWinters(ts_data)
plot(model)
```
